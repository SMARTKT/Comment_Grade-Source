```python
import pandas as pd
import pickle
import numpy as np
```

```python
CORE_RESPONSES = pd.read_csv('all_results_categories_raw.csv',delimiter='\t')
columns = CORE_RESPONSES.columns
CORE_RESPONSES.head()

df_lib = CORE_RESPONSES['Filename'].str.contains("PLplot") 

# curl mariadb  dealii PLplot
data = CORE_RESPONSES[df_lib]
data.to_csv('libpng_data_features.csv')
```

```python
THRESHOLD = np.zeros(len(columns))+ 3

print(len(columns))
THRESHOLD[4]=2
THRESHOLD[16]=1.5
THRESHOLD[17]=0.5
THRESHOLD[18]=1.5
print (THRESHOLD)
# Change threshold of corresponding column here, eg. THRESHOLD[4] = 6 (num problem domain)
```

```python
cats_count =[]
data_np = np.array(data)
cats = []
for i in range(len(data_np)):
    temp = []
    for j in range(3,len(columns)):
        if data_np[i,j] >= THRESHOLD[j]:
            temp.append(columns[j])
            cats_count.append(columns[j])
    cats.append(','.join(temp))
```

```python
cats_count
```

```python
data['Categories Matches'] = cats
```

```python
data.to_csv('all_results_categories_libpng_new_srijoni.csv',sep='\t',index=None)
```
