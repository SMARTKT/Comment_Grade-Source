```python
import os
import pandas as pd
import numpy as np
from os.path import join as PJOIN
from collections import Counter
import xlsxwriter as xl
```

```python
ANNOTATIONS_FILE1 = "DATA/GENERATED_NEW/server/MariaDB_Yash_1900.xlsx"
ANNOTATIONS_FILE2 = "DATA/GENERATED_NEW/server/Subhanan_Last_500_mariaDB.xlsx"

def filenames(file):
    leftovername = file[len("DATA/GENERATED_NEW/"):].replace("/","_")
    return leftovername
    #outputfilename = "ANALYSIS/DIFF/"+leftovername
FNAME1 = filenames(ANNOTATIONS_FILE1)
FNAME2 = filenames(ANNOTATIONS_FILE2)

OUTPUT_FILENAME = "ANALYSIS/BIAS/" + FNAME1 + "__" + FNAME2
    
# PROJECT_NAME = "OpenCPN"
# AUTHOR_NAME = "allmarked"
# OUTPUT_DIR = "DATA/GENERATED/TRAIN/"
# OUTPUT_SUFFIX = "_cal"
# OUTPUT_FILENAME = "train_" + PROJECT_NAME +"_" + AUTHOR_NAME + OUTPUT_SUFFIX + ".csv"
ANNOTATIONS = {'N':1,'P':2,'U':3,'n':1}
REVERSE_MAP = {1: 'N', 2: 'P', 3:'U'}
```

```python
def read_file(file):
    if file[-3:] == 'csv':
        annotations_file = pd.read_csv(file,delimiter='$')
    else:
        annotations_file = pd.read_excel(file)
    return annotations_file
annotations_file1 = read_file(ANNOTATIONS_FILE1)
annotations_file2 = read_file(ANNOTATIONS_FILE2)
```

```python
annotations_np1 = np.array(annotations_file1)
annotations_np2 = np.array(annotations_file2)
```

```python
for i,el in enumerate(annotations_np1[500]):
    print(i," : ",el)
```

```python
for i,el in enumerate(annotations_np2[0]):
    print(i," : ",el)
```

```python
INTUITIVE_INDEX1 = 37
CALCULATED_INDEX1 = 38

INTUITIVE_INDEX2 = 36
CALCULATED_INDEX2 = 37

CATEGORIES_START1 = 6
CATEGORIES_START2 = 5
```

```python
def index(file_data):
    mp = {}
    for el in file_data:
        key = (el[0], el[1])
        mp[key] = el
    return mp

comment_data1 = index(annotations_np1)
comment_data2 = index(annotations_np2)
```

```python
wb = xl.Workbook(OUTPUT_FILENAME)
ws = wb.add_worksheet()

row = 0

def write_row(rn, data, ws):
    for j,el in enumerate(data):
        if el != el:
            continue
        ws.write(rn,j,el)

for k,v1 in comment_data1.items():
    if k not in comment_data2:
        continue
    #print("HERE")
    v2 = comment_data2[k]
    labels_intuitive1 = v1[INTUITIVE_INDEX1]
    labels_intuitive2 = v2[INTUITIVE_INDEX2]
    
    labels_calculated1 = v1[CALCULATED_INDEX1]
    labels_calculated2 = v2[CALCULATED_INDEX2]
    
    rules1 = v1[CALCULATED_INDEX1+1]
    rules2 = v2[CALCULATED_INDEX2+1]
    
    if (labels_intuitive1 != labels_intuitive1) or (labels_intuitive2 != labels_intuitive2):
        continue
    if (labels_calculated1 != labels_calculated1) or (labels_calculated2 != labels_calculated2):
        print("Calculated Label not present")
        print(v1)
        print(v2)
    towrite1 = [v1[0], v1[1]]
    towrite1.extend(v1[CATEGORIES_START1:CATEGORIES_START1+31])
    towrite1.extend([labels_intuitive1, labels_calculated1, rules1, FNAME1])
    
    towrite2 = [v2[0], v2[1]]
    towrite2.extend(v2[CATEGORIES_START2:CATEGORIES_START2+31])
    towrite2.extend([labels_intuitive2, labels_calculated2, rules2, FNAME2])
    
    write_row(row, towrite1, ws)
    row += 1
    write_row(row, towrite2, ws)
    row += 1
    
wb.close()
```
